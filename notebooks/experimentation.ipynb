```python
import sys
import os
import os
import pandas as pd
import joblib
from sklearn.metrics import accuracy_score
# Ajouter le chemin du r√©pertoire parent de cifar10_classification
project_path = os.path.abspath('..')
if project_path not in sys.path:
    sys.path.append(project_path)
print(project_path)
```

```python
def save_model(model, feature_method, classifier):
    model_dir = '/home/mkbrad7/afs_epita/ING2/ML_reconnaissance_de_forme/Projet/classifiaction_cifar/models/'
    #os.makedirs(model_dir, exist_ok=True)
    model_filename = os.path.join(model_dir, f'{classifier}_{feature_method}_model.pkl')
    joblib.dump(model, model_filename)

    # model_name = f'{feature_method}_{classifier}.joblib'
    # joblib.dump(model, os.path.join(model_dir, model_name))
```

```python
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
from cifar10_classification.dataset import prepare_data
from cifar10_classification.features import extract_features
from cifar10_classification.modeling.train import train_classifier
from cifar10_classification.config import LABEL_NAMES, MODEL_DIR
from cifar10_classification.plots import plot_confusion_matrix, plot_roc_curve

X_train, y_train, X_val, y_val, X_test, y_test = prepare_data()

feature_methods = ['flatten', 'hog']
classifiers = ['logistic', 'random_forest', 'sgd'] # 'svm'
results = []
for feature_method in feature_methods:
    X_train_features = extract_features(X_train, method=feature_method)
    X_val_features = extract_features(X_val, method=feature_method)
    X_test_features = extract_features(X_test, method=feature_method)

    for classifier in classifiers:
        print(f"Training {classifier} with {feature_method} features...")
        model, val_predictions = train_classifier(X_train_features, y_train, X_val_features, y_val, model_type=classifier)
        
        # Evaluate on validation set
        report = classification_report(y_val, val_predictions, target_names=LABEL_NAMES, output_dict=True)
        accuracy = accuracy_score(y_val, val_predictions)
        results.append({'feature_method': feature_method, 'classifier': classifier, 'report': report, 'accuracy': accuracy})
        print(results[-1]['report'])

        # Save the model
        save_model(model, feature_method, classifier)
        # Confusion matrix
        cm = confusion_matrix(y_val, val_predictions)
        plot_confusion_matrix(cm, classes=LABEL_NAMES)

        # ROC curve
        if classifier in ['logistic', 'sgd']:
            y_val_prob = model.predict_proba(X_val_features)
            plot_roc_curve(y_val, y_val_prob, n_classes=len(LABEL_NAMES))

    # Save results to a DataFrame and then to an Excel file
    results_df = pd.DataFrame(results)
    results_df.to_excel('model_results.xlsx', index=False)
```
