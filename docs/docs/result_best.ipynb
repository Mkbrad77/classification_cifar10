```python
import pandas as pd

# Charger le fichier Excel
file_path = '../model_results_2best.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Filtrer les lignes avec 'feature_method' égal à 'sift' et 'classifier' égal à 'random_forest' ou 'logistic'
filtered_df = df[(df['feature_method'] == 'sift') & 
                 ((df['classifier'] == 'random_forest') | 
                  (df['classifier'] == 'logistic'))]

# Afficher le DataFrame filtré
print(filtered_df)
```

```python
filtered_df
```

```python
# Charger le nouveau fichier Excel où insérer les données filtrées
file_path_new = '../model_results_best.xlsx'
df_new = pd.read_excel(file_path_new, sheet_name='Sheet1')
```

```python
df_new
```

```python
# Ajouter les données filtrées au DataFrame existant
df_combined = pd.concat([df_new, filtered_df], ignore_index=True)
```

```python
df_combined
```

```python


# Enregistrer le DataFrame combiné dans un nouveau fichier Excel
df_combined.to_excel('../all_model_results.xlsx', index=False)
```
